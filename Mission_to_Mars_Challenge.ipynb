```python
# Import Splinter and BeautifulSoup
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the mars nasa news site
url = 'https://redplanetscience.com'
browser.visit(url)
# Optional delay for loading the page
browser.is_element_present_by_css('div.list_text', wait_time=1)
```

```python
html = browser.html
news_soup = soup(html, 'html.parser')
slide_elem = news_soup.select_one('div.list_text')
```

```python
slide_elem.find('div', class_='content_title')
```

```python
# Use the parent element to find the first `a` tag and save it as `news_title`
news_title = slide_elem.find('div', class_='content_title').get_text()
news_title
```

```python
# Use the parent element to find the paragraph text
news_p = slide_elem.find('div', class_='article_teaser_body').get_text()
news_p
```

```python
# Visit URL
url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(url)
```

```python
image_urls = []

# retrieve full resolution image urls for each hemisphere
full_image = browser.find_by_tag("h3")
for item in range(len(full_image)):
    
    # add info into empty dictionary for storage
    hemispheres = {}
    
    #Click on the item in h3
    browser.find_by_tag("h3")[item].click()
    
    #complete url from <a tag wit 'original' text
    #download_url = browser.find_by_tag("h3")
    #donwload_url = browser.find_by_css(dd['href'])

    download_url = browser.find_by_text("Sample")
    
    #add img url to dictionary
    hemispheres["img_url"] = download_url["href"]
    
    #title from heading and add to dictionary
    hemispheres["title"] = browser.find_by_tag("h2").text
    
    # append urls to hemispheres list
    #image_urls.append(hemispheres)
    image_urls.append(hemispheres)

    
    browser.back()
```
